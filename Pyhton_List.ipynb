Problem Statement:
Given a list of numbers, find the sum of all pairs (i, j) such that

both i and j belong to the list

and i <= j

Store the final sum inside a list and print it.

(Your code clearly tries to compare two elements and add something to sum.)

```python
l = [1, 2, 3, 4, 5, 4]
result = []

for i in l:
    total = 0          # reset for each i
    for j in l:
        if i <= j:
            total = total + j
    result.append(total)

print(result)
```

Problem statement :
You are given a list of strings that may contain alphabets and digits mixed together.

L = ['1ac21', '23fg', '456', '098d', '1', 'kls']

Task:

For each string in the list:

Extract only the digits.

Calculate the product of those digits.

If a string has no digits, its product should be considered 1 (default initialization).

Create a list that stores the product corresponding to each string.

Sort the original strings in descending order based on their calculated digit-products.

```python
# Write code here
L = ['1ac21', '23fg', '456', '098d', '1', 'kls']
prod_value = []

for item in L:
    product = 1
    for char in item:
        if char.isdigit():
            product = product * int(char)
    prod_value.append(product)

# sorting strings based on product value (descending)
result = [i[1] for i in sorted(list(zip(prod_value, L)), reverse=True)]

print(result)
```
