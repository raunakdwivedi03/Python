TUPLEs in Python

A tuple in Python is a collection of items that is:

Ordered (items have a fixed position/index)

Immutable (cannot be changed after creation)

Can store different data types

```python
# 1Ô∏è‚É£ Empty tuple
t1 = ()
print(t1)            # Output: ()

# 2Ô∏è‚É£ Single element tuple (comma is required)
t2 = ('hello',)
print(t2)            # Output: ('hello',)
print(type(t2))      # Output: <class 'tuple'>

# 3Ô∏è‚É£ Homogeneous tuple (same data type elements)
t3 = (1, 2, 3, 4)
print(t3)            # Output: (1, 2, 3, 4)

# 4Ô∏è‚É£ Heterogeneous tuple (different data types)
t4 = (1, 2.5, True, [1, 2, 3])
print(t4)            # Output: (1, 2.5, True, [1, 2, 3])

# 5Ô∏è‚É£ Nested tuple (tuple inside another tuple)
t5 = (1, 2, 3, (4, 5))
print(t5)            # Output: (1, 2, 3, (4, 5))

# 6Ô∏è‚É£ Tuple using type conversion
t6 = tuple('hello')
print(t6)            # Output: ('h', 'e', 'l', 'l', 'o')
```

```python
# DELETING ELEMENTS FROM A SET

#  remove()
# Removes a specific element from the set
# If the element does NOT exist ‚Üí ERROR occurs
S1 = {1, 2, 3, 4}
S1.remove(4)
print(S1)


#  discard()
# Removes a specific element from the set
# If the element does NOT exist ‚Üí NO error
S1 = {1, 2, 3, 4}
S1.discard(4)
print(S1)


#  pop()
# Removes and returns a RANDOM element from the set
# Because sets are unordered, you don't know which element is removed
S1 = {1, 2, 3, 4}
removed_element = S1.pop()
print(removed_element)  # Random value
print(S1)               # Remaining elements


#  clear()
# Removes ALL elements from the set
# Set becomes empty
S1 = {1, 2, 3, 4}
S1.clear()
print(S1)
```

SET operations

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

# -------------------------------
# UNION ( | )
# Combines elements from both sets
# Removes duplicates automatically
print(s1 | s2)
# Output: {1, 2, 3, 4, 5, 6, 7, 8}


# -------------------------------
# INTERSECTION ( & )
# Returns only common elements between both sets
print(s1 & s2)
# Output: {4, 5}


# -------------------------------
# DIFFERENCE ( - )
# Elements present in s1 but NOT in s2
print(s1 - s2)
# Output: {1, 2, 3}

# Elements present in s2 but NOT in s1
print(s2 - s1)
# Output: {6, 7, 8}


# -------------------------------
# SYMMETRIC DIFFERENCE ( ^ )
# Elements that are in either set
# BUT NOT common in both
print(s1 ^ s2)
# Output: {1, 2, 3, 6, 7, 8}


# -------------------------------
# MEMBERSHIP TEST
# Checks whether an element exists in the set
print(1 in s1)     # True
print(10 in s1)    # False


# -------------------------------
# ITERATION
# Looping through a set
for i in s1:
    print(i)
```

Dictionary ->A dictionary in Python is a built-in data type used to store data in key‚Äìvalue pairs.

üîπ Simple Definition

A dictionary stores data like a real-life dictionary:

word (key) ‚Üí meaning (value)

```python
#dict_name = {key: value, key: value}
```

```python
#mutable
#indexing has no meaning
#keys can't  be duplicated
#keys can't be mutable items
```

```python
# 1. Empty Dictionary
d1 = {}
print("Empty Dictionary:", d1)

# 2. 1D Dictionary
d2 = {
    "name": "Amit",
    "age": 21,
    "course": "Python"
}
print("1D Dictionary:", d2)

# 3. Dictionary with Mixed Keys
d3 = {
    1: "one",
    "two": 2,
    (3, 4): "tuple key"
}
print("Mixed Keys Dictionary:", d3)

# 4. 2D Dictionary (Nested Dictionary) --> JSON
d4 = {
    101: {"name": "Rahul", "age": 20},
    102: {"name": "Anita", "age": 21}
}
print("2D Dictionary:", d4)

# 5. Using Sequence and dict() function
pairs = [(1, "a"), (2, "b"), (3, "c")]
d5 = dict(pairs)
print("Using sequence:", d5)

# Using zip()
keys = ["name", "age", "city"]
values = ["Rohit", 22, "Delhi"]
d6 = dict(zip(keys, values))
print("Using zip():", d6)

# 6. Duplicate Keys
d7 = {
    "a": 1,
    "b": 2,
    "a": 3
}
print("Duplicate Keys Dictionary:", d7)

# 7. Mutable Items as Keys (NOT ALLOWED)
# This will give an error because list is mutable
# d8 = {[1, 2, 3]: "list key"}

# Correct way (use tuple instead)
d8 = {(1, 2, 3): "tuple key"}
print("Immutable Key Dictionary:", d8)
```

REmoving key value pairs


```python
# Correct Dictionary
d = {
    'name': 'Rooney',
    'age': 32,
    3: 3,
    'gender': 'male',
    'weight': 72
}

print("Original Dictionary:", d)

# pop() ‚Üí removes specific key
d.pop('age')
print("After pop('age'):", d)

# popitem() ‚Üí removes last inserted item
d.popitem()
print("After popitem():", d)

# del ‚Üí deletes specific key
del d['name']
print("After del['name']:", d)

# clear() ‚Üí removes all items
d.clear()
print("After clear():", d)
```

```python
squares = {x: x*x for x in range(1, 6)}
print(squares)
```

```python
prices = {"apple": 100, "banana": 40, "mango": 80}

discount = {k: v-10 for k, v in prices.items()}
print(discount)
```

```python
#using zip
keys = ["name", "age", "city"]
values = ["Rooney", 32, "London"]

d = {k: v for k, v in zip(keys, values)}
print(d)
```
