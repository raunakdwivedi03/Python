```python
n = int(input("Enter the number: "))
total = 0

for i in range(1, n + 1):
    if i % 5 == 0:
        continue   # skip multiples of 5

    total += i     # add non-multiples of 5

    if total > 300:
        total -= i  # undo last addition
        break

print("Sum of the numbers is:", total)
```

```python
##write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers
total = 0
count = 0

while True:
    num = int(input("Enter the number: "))

    if num == 0:
        break

    total += num
    count += 1

if count > 0:
    print("Sum:", total)
    print("Average:", total / count)
else:
    print("No numbers entered")
```

```python
#printing Prime number
l = int(input('enter the lower limit: '))
u = int(input('enter the upper limit: '))

for i in range(l, u + 1):
    if i > 1:
        is_prime = True
        for j in range(2, i):
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            print(i)
```
