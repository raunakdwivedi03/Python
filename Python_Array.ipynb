```python

##1. find the kth largest and smallest items from a list
l=[12,32,4,42,12,1,3,4,2,4,2,43,2]
k=3
l.sort()
print(l)

print(l[k-1])##minimum
print(l[-k]) ## maximum

```

```python
#3. find min/max in given array
l=[3,2,4,34,2,4,24,2,4,3]
print(max(l))
print(min(l))
```

```python
#optimal approach
l=[3,2,4,34,2,4,24,2,4,3]
max_val=l[0]
min_val=l[0]

for i in l:
  if i>max_val:
    max_val=i
  if i<min_val:
    min_val=i

print(max_val)
print(min_val)
```

```python
#4. maximum sum subarray
l=[-2,4,7,-1,6,-11,14,3,-1,-6]

d={}
#brute force approach
for i in range(0,len(l)):
  subarray=[]
  for j in range(i,len(l)):
    subarray.append(l[j])
    d[sum(subarray)]=subarray

max_val = max(d.keys())

for i in d:
  if i==max_val:
    print(d[i])
```

```python
l = [2, -3, 4, -1, 2, 1, -5, 4]

curr_sum = 0
curr_seq = []

best_sum = float('-inf')
best_seq = []

for i in l:                      # âœ… FIXED
    if curr_sum + i > i:
        curr_sum = curr_sum + i
        curr_seq.append(i)
    else:
        curr_sum = i
        curr_seq = [i]

    if curr_sum > best_sum:
        best_sum = curr_sum
        best_seq = curr_seq.copy()

print("Best Subarray:", best_seq)
print("Max Sum:", best_sum)
```

```python
##find the element with left side smaller and right side greater in the array
l = [3,1,2,5,6,8,7,9]

for i in range(1, len(l)-1):
    flag = True

    for j in range(0, i):
        if l[j] > l[i]:
            flag = False
            break

    for j in range(i+1, len(l)):
        if l[j] < l[i]:
            flag = False
            break

    if flag:
        print(l[i])
```

```python
##optimal  approach
l=[3,1,2,5,6,8,7,9]

for i in range(1,len(l)-1):
  if max(l[0:i])<l[i] < min(l[i+1:]):
    print(l[i])
```

```python
l = [3,1,2,5,8,7,9]

max_arr = []
min_arr = []

max_val = l[0]
min_val = l[-1]

# Prefix maximum array
for i in l:
    if i > max_val:
        max_val = i
    max_arr.append(max_val)

# Suffix minimum array
for i in range(len(l)-1, -1, -1):
    if l[i] < min_val:
        min_val = l[i]
    min_arr.insert(0, min_val)

# Find valid elements
for i in range(1, len(l)-1):
    if max_arr[i-1] < l[i] < min_arr[i+1]:
        print(l[i])

print("min_arr:", min_arr)
print("max_arr:", max_arr)
```

```python
##find the continuos subarray with a given sum(given non negative numbers )
# return the starting and ending index of the subarray
# return 1st subarray in case of multiple

l=[1,22,13,7,9,11,10]
s=16

for i in range(0,len(l)):
  subarray=[]
  for j in range(i+1,len(l)):
    subarray.append(l[j])
    if sum(subarray)==s:
      print(subarray)
```

```python
##optimal approach
a=[1,2,3,4,5,6,7,7]
b=[3,4,5,6,7,8,9]

i =j=0
while i<len(a) and j<len(b):
  if a[i]==b[j]:
    print(a[i])
    i+=1
    j+=1
  elif a[i]>b[j]:
    j+=1
  else:
    i+=1

```

```python
##rotate the array to left d items
a=[1,2,3,4,5,6,7,7]
rotate=3

for i in range(0,rotate):
  a.append(a[0])
  a.pop(0)

print(a)
```

```python
##optimal approach
l=[1,2,3,4,5,6,7,7]
rotate=2

for i in range(rotate):
  temp=l[0]
  for j in range(0,len(l)-1):
    l[j]=l[j+1]
  l[len(l)-1]=temp

print(l)
```

```python
#find the duplicate in an array
a=[1,1,2,3,4,5,6,7,7]

result=[]
for i in a:
  if a.count(i)>1 and i not in result:
    result.append(i)

print(result)
```
