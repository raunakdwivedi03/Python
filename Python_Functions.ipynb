What is *args?

*args allows a function to accept multiple positional arguments.

ðŸ“Œ args = arguments (name kuch bhi ho sakta hai)

ðŸ”¹ What is **kwargs?

**kwargs allows a function to accept multiple keyword arguments.

ðŸ“Œ kwargs = keyword arguments

```python
##args

def multiply(*args):
  product=1

  for i in args:
    product=product*i

  print(args)
  return product
```

```python
##**kwargs
##they contains key value pairs like python dictionary

def display(**kwargs):

  for key,value in kwargs.items():
    print(key,'->',value)
```

Lambda Function in Python ðŸ

A lambda function is a small, anonymous (no-name) function in Python.
It is used when you need a short function for a short time.

```python
## lambda functions
lambda arguments: expression
```

```python
#x--x^2
a=lambda x:x**2
a(4)
```

```python
#x,y-->x+y
a=lambda x,y:x+y
a(2,3)
```

Difference between lambda functions and Normal functions

Lambda function = chhota, one-line, temporary function

ðŸ‘‰ Normal function = full-feature, reusable, readable function

```python
#check if a string has a
a= lambda s: 'a' in s
a('hello')
```

```python
##odd and even
a=lambda x:' even ' if x%2==0 else 'odd'
a(6)
```

lamda are used in Higher order functions

```python
# normal function
def square(x):
    return x ** 2

# Higher Order Function
def transform(f, l):
    output = []
    for i in l:
        output.append(f(i))
    return output

# list
l = [2, 3, 1, 3, 42, 4]

# function call
result = transform(lambda x:x**4, l)
print(result)
```
