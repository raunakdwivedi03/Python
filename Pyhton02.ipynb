```python
##count the frequency of particular frequncy in given qord
s = input('enter the string: ')
term = input('what would like to search for: ')

count=0
for i in s:
  if i==term:
    count+=1

print('frequency ' , count)
```

```python
##write the program to remove the particular charcter in a string
s = input('enter the string: ')
term = input('what would like to search for: ')

result=''

for i in s:
  if i != term:
    result = result + i

print(result)
```

```python
## write a program that can check whether is string is palindrome or not
#ex- abba , malyalam

s = input('enter the string: ')
flag=True
for i in range(0,len(s)//2):
  if s[i]!=s[len(s) - i -1]:
    flag=False
    print('Not palindrome')
    break


if flag:
  print('palindrome')
```

```python
## write a program to count the number of words in a string without using the split()

s=input('enter the string: ')
l=[]
temp=''

 for i in s:

  if i != ' ':
    temp=temp+i
  else:
    l.append(temp)
    temp=''

l.append(temp)
print(l)
```

```python
###write the python program to convert a string to tittle case without using the tittle()
s=input('enter the string: ')
l=[]

for i in s.split():
  l.append(i[0].upper()+ i[1:].lower())

print("".join(l))
```
